```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import Packages we are going to use.
import numpy as np
import pandas as pd
```

```python
#Read CSV file in a Dataframe with proper encoding
rounds2 = pd.read_csv("rounds2.csv", sep=None) 
print(rounds2.shape)#Print Shape to get the size of the DataFrame

##Convert the Permalink column to utf-8 for this we first encode with UTF-8 and decode using ASCII
rounds2.company_permalink = rounds2.company_permalink.str.encode('UTF-8').str.decode('ASCII', 'ignore')
rounds2.company_permalink = rounds2.company_permalink.str.encode('UTF-8').str.decode('ASCII', 'ignore')
rounds2 # Print the file to verify
```

```python

#Import text File into Dataframes
companies = pd.read_csv("companies.txt",sep=None)
print("Shape of Dataframe:")
print(companies.shape) #Print Shape to get the size of the DataFrame

##Convert the company_permalink column to utf-8 again the same first encode with utf-8 and then decode with ASCII
companies.permalink = companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
companies.head(100)#Print first 100 rows from DataFrame
```

```python
#Convert both the Dataframe's first column into lower case for comparing.
rounds2["company_permalink"] = rounds2["company_permalink"].map(lambda x: x.lower()) 
companies["permalink"] = companies["permalink"].map(lambda x: x.lower())

#To check whether there are any companies that are in round2 but not in Companies
#No Rows should be returned from below statement.
rounds2.loc[~rounds2['company_permalink'].isin(companies['permalink']), :]
```

```python
rounds2 = rounds2.drop_duplicates(subset='funding_round_permalink') #Remove Duplicates from rounds2 Dataframe
companies = companies.drop_duplicates(subset='permalink') #Remove Duplicates from companies Dataframe
print(rounds2.shape)
print(companies.shape)
```

```python
#Column Wise sum of Null values in both the DataFrames
print("Percentage of Null values in Round2 Dataframe: ")
print(round(100*(rounds2.isnull().sum()/len(rounds2)), 2)) #Print Percentage of Null Values in round2 Dataframe
print(len(rounds2)) #Total Length or No. of rows in rounds2 Dataframe

print("\n\nPercentage of Null Values in Companies Dataframe:")
print(round(100*(companies.isnull().sum()/len(companies)), 2)) #Print percentage of Null Values from Companies Dataframe
print(len(companies)) #Total Length or No. of rows in Companies Dataframe
```

```python
print("\n\nPercentage of Null Values in Companies Dataframe:")
print(round(100*(companies.isnull().sum()/len(companies)), 2)) #Print percentage of Null Values from Companies Dataframe
print(len(companies)) #Total Length or No. of rows in Companies Dataframe
```

```python
master_frame = pd.merge(rounds2, companies, how = 'left', left_on = 'company_permalink', right_on = 'permalink')
print(master_frame.shape)
master_frame
```

```python
print("Percentage of Null values in master_frame Dataframe: ")
print(round(100*(master_frame.isnull().sum()/len(master_frame)), 2)) 
```

```python
#Since many of the columns which has null values are not required in analysis so we should drop them.

columns = ['funding_round_code', 'founded_at', 'city', 'region', 'state_code', 'status', 'homepage_url', 'funded_at']
master_frame = master_frame.drop(columns, axis=1)

print("Percentage of Null values in master_frame Dataframe: ")
print(round(100*(master_frame.isnull().sum()/len(master_frame)), 2)) 
```

```python
#We first drop the values which are Null in raised_amount_usd movies = movies[~np.isnan(movies['gross'])]
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]

#Now we impute the category_list null values with others we put unknowns in others list.
master_frame['category_list'].fillna('others', inplace=True)
```

```python
#we chose to impute the values for Country as the % of rows are 8% and we have already lost about 22% of data,
#So I think its better to impute the values
#For this we finds the distrbution of the various variables in the category_list 
sample_distn = master_frame.country_code.value_counts(normalize = True) 
NA_values = master_frame.country_code.isnull()#Finds the null values
#And replaces the null values WRT the Distribution
master_frame.loc[NA_values,'country_code'] = np.random.choice(sample_distn.index, size = 
                                                              len(master_frame[NA_values]),p = sample_distn.values) 

print("Percentage of Null values in master_frame Dataframe: ")
print(round(100*(master_frame.isnull().sum()/len(master_frame)), 2))

print("Percentage of Rows retained in master_frame Dataframe: ")
print(round(100*(len(master_frame) / 114949), 2))#66368 is the original dataframe row count after merging two columns
master_frame
```

```python
#Convert Fund into Million Dollars and save in a new column.
master_frame['mDollars'] = (master_frame.raised_amount_usd/1000000)
master_frame
```

```python
print("Average of Fund Raised in million $: ")
print(master_frame['mDollars'].groupby(master_frame['funding_round_type']).mean().sort_values(ascending=False))

print("\n\nAverage of Fund Raised in $: ")
print(master_frame['raised_amount_usd'].groupby(master_frame['funding_round_type']).mean().
      round().sort_values(ascending=False))

##By Looking at the result we can very well make out that Spark Fund should go for Venture Category since they are looking
# looking to invest bertwen $5m to $15m
```

```python
##Print Countries where most Investments have taken place for Venture Funding Round in Million Dollars
print(master_frame['mDollars'].loc[master_frame['funding_round_type'] == 'venture'].groupby(master_frame['country_code'])
      .sum().round().sort_values(ascending=False))
```

```python
#Filter out top 9 countries of Funding Round named as Venture
top9 = master_frame['mDollars'].loc[master_frame['funding_round_type'] == 'venture'
                        ].groupby(master_frame['country_code']).sum().round().sort_values(ascending=False).head(9)
top9.shape#To Verify if we have got top 9 rows
top9 #Print the new DataFrame
```

```python
#So if we have a look at the Above Table we can make out that the Top Three English Speaking Country with highest
#Venture Funding are 
#1. USA
#2. GBR
#3. IND
```

```python
#Add a Primary Category Column in master_frame by splitting the first category from category list separated by |
master_frame['primarySector'] = master_frame['category_list'].str.split('|').str[0]
master_frame#To verify
```

```python
#Read CSV file in a Dataframe with proper encoding
mapping = pd.read_csv("mapping.csv", sep=None) 
print(mapping.shape)#Print Shape to get the size of the DataFrame

#Replace unwanted character 0 with na like in Waste Management it was Waste Ma0gement
mapping['category_list'] = mapping['category_list'].str.replace('0', 'na')
mapping # Print the file to verify
```

```python
meltedMapp = pd.melt(mapping, id_vars=["category_list"],var_name="MainSector") #Use the Melt function to reshape Mapping
meltedMapp # To verify
```

```python
#The melt function has column to rows for each rows that is why we have 6192 rows but we have a value column to check
#if the melting is done properly for the row or not, if not, we can drop the row.
meltedMapp = meltedMapp[meltedMapp['value'] > 0]#Take only rows which matters
print(meltedMapp.shape)#Check the shape
meltedMapp # To verify
```

```python
#Merge with master_frame to obtain the desired result with primary Category on the right most column
master_frame = pd.merge(master_frame, meltedMapp, how = 'inner', left_on = 'primarySector', right_on = 'category_list')
master_frame#To verify
```

```python
##We can Drop some unwanted columns from the Dataframe to make it look better
columns = ['category_list_y', 'value']
master_frame = master_frame.drop(columns, axis=1)
master_frame.rename(columns = {'category_list_x':'category_list'}, inplace = True)
master_frame #To verify
```

```python
#Find the Sum and count of Main categories in a separate column Sum is in Million Dollars
master_frame['sumMainCat'] = master_frame.groupby('MainSector')['mDollars'].transform('sum')
master_frame['countMainCat'] = master_frame.groupby('MainSector')['MainSector'].transform('count')
```

```python
master_frame #Verify if the two Columns are added
```

```python
#Let us define Tuples for Country code since tuples are immutable Data Structure
country = ('USA', 'GBR', 'IND')

#We define the Funding Type as FT as well
FT = ('venture')
#To verify
print(country)
print(FT)
```

```python
#DataFrame for Top Most English Speaking Country
D1 = master_frame.loc[(master_frame['funding_round_type'] == FT) & (master_frame['country_code'] == country[0])
                     & (master_frame['mDollars'].between(5, 15, inclusive=True))]
D1 #Check Data Frame
```

```python
#Dataframe for Second highest funded English Speaking Country
D2 = master_frame.loc[(master_frame['funding_round_type'] == FT) & (master_frame['country_code'] == country[1])
                     & (master_frame['mDollars'].between(5, 15, inclusive=True))]
D2
```

```python
#Dataframe for Third highest funded English Speaking Country with Best investment between $5 - $15M
D3 = master_frame.loc[(master_frame['funding_round_type'] == FT) & (master_frame['country_code'] == country[2])
                     & (master_frame['mDollars'].between(5, 15, inclusive=True))]
D3
```

```python
#Total amount raised by comapnies in USA, GBR and India in that order
countrySum = {'Country' : ['USA', 'GBR', 'IND'],
              'AmountSum' : [round(D1.mDollars.sum(), 2), round(D2.mDollars.sum(), 2), round(D3.mDollars.sum(), 2)]
             }
countrySum = pd.DataFrame(countrySum)
countrySum
```

```python
#Sizes of DF's D1, D2 and D3 in that order
countryCount = {'Country' : country,
              'AmountSum' : [D1.shape[0], D2.shape[0], D3.shape[0]]
             }
countryCount = pd.DataFrame(countryCount)
countryCount
```
